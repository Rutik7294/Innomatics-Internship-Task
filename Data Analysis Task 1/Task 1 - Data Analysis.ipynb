# Fetching important columns for analysis

```python
imp_cols = [
    "Salary",
    "10percentage",
    "12percentage",
    'CollegeTier',
    'Degree',
    'collegeGPA',
    'English',
    'Logical',
    'Quant',
    'Domain',
    'ComputerProgramming',
    'ElectronicsAndSemicon',
    'ComputerScience',
    'MechanicalEngg',
    'ElectricalEngg',
    'TelecomEngg',
    'CivilEngg'
]
```

```python
new_data = data[imp_cols]
```

## correlation for the new dataframe

```python
plt.figure(figsize = (15, 7))
sns.heatmap(correlation)
plt.show()
```

## The DataFrame for computer science students

```python
comp_data = new_data[(new_data['ComputerProgramming'] > -1) & (new_data['ComputerScience'] > -1)]
```

```python
comp_data.head()
```

```python
comp_data['Degree'].unique()
```

## Describing the statistics

```python
comp_data.describe()
```

## Data Visualization Phase

```python
sns.set(font_scale = 2)

plt.figure(figsize = (25, 15))
plt.title("Salaries according to the Degrees of students")
sns.boxplot(x = 'Degree', y = 'Salary', data = comp_data.sort_values('Salary', ascending = False))
```

```python
sns.set(font_scale = 2)

plt.figure(figsize = (25, 15))
plt.title("Salaries according to the Degrees of students")
sns.lineplot(x = 'Degree', y = 'collegeGPA', data = comp_data.sort_values('collegeGPA', ascending = False))
```

```python
sns.set(font_scale = 2)
plt.figure(figsize = (14, 7))
plt.title("The Score of students for Computer Programming")
plt.xlabel("Degree")
plt.ylabel("Score")
plt.bar(comp_data['Degree'], comp_data['ComputerProgramming'], color = 'red')
plt.show()
```

```python
plt.figure(figsize = (14, 7))
plt.title("The Score of students for Quantitative Aptitude")
plt.xlabel("Degree")
plt.ylabel("Score")
plt.bar(comp_data['Degree'], comp_data['Quant'], color = 'green')
plt.show()
```

```python
plt.figure(figsize = (14, 7))
plt.title("The Score of students for Logical Reasoning")
plt.xlabel("Degree")
plt.ylabel("Score")
plt.bar(comp_data['Degree'], comp_data['Logical'], color = 'pink')
plt.show()
```

```python
plt.figure(figsize = (14, 7))
plt.title("The Score of students for English Speaking")
plt.xlabel("Degree")
plt.ylabel("Score")
plt.bar(comp_data['Degree'], comp_data['English'], color = 'orange')
plt.show()
```
